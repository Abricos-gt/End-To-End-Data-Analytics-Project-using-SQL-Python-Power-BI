```python
# fill the missing value of each category by the median of the category it belongs
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# change the name of attributes in to snake case
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group
```

```python
labels = ['Young Adult', 'Adult' , 'Middle_Aged','Senior'] 
df['age_group']= pd.qcut(df['age'] , q = 4, labels = labels)
```

```python
df[['age','age_group']].head()
```

```python
df['age_group'].value_counts()
```

```python
# Create a column of purchase_frequency_ days
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Annually':365,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Monthly':30,
    'Every 3 Months':90
}
```

```python
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping )
```
